```python
import os
import sys
sys.path.insert(0, os.path.abspath("../../../gpsp-atlas"))

import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from gpsp_atlas_helper import GPSPAtlasHelper
import numpy as np
from matplotlib.ticker import LogLocator, MultipleLocator, NullFormatter, LogFormatterSciNotation
import matplotlib.patheffects as path_effects
```

```python
plt.rcParams.update({
    'font.size': 12,
    'axes.labelsize': 16,
    'axes.titlesize': 16,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'legend.fontsize': 16,
    'figure.titlesize': 18,
    'axes.linewidth': 1.2     
})
```

```python
# --- CONFIGURATION ---
atlas_file_R12 = "/home/oguzhan/Python3Workspace/Survival_Probability_Atlas/PaperCodes/GPSP_Atlas_4D_R12_20bins.fits"
#atlas_file_F98 = "GPSP_Atlas_4D_F98_20bins.fits"
# Define Galactic Center region (l,b) = (0.0)
target_l, target_b = 0.0, 0.0
# Define distance of 8.5 kpc
target_d = 8.5 # kpc
# Define energy of 2 PeV and 150 TeV
target_e = 2e15 # Energy in eV
target_e2 = 1.5e14     # 1 PeV in eV
```

```python
atlas = GPSPAtlasHelper(atlas_file_R12)
fig = plt.figure(figsize=(8, 6))
(X, Y), Z = atlas.get_2d_phase_space(target_l, target_b)
im1 = plt.pcolormesh(X, Y, Z, cmap='cividis', shading='auto', norm=Normalize(0, 1.0), rasterized=True, edgecolor='face')

#Define survival probability contour levels here, levels=[0.1, 0.37, 0.85]
cs = plt.contour(X, Y, Z, levels=[0.1, 0.37, 0.85], colors='white', linestyles='--', linewidths=1.5)
plt.setp(cs.collections, path_effects=[path_effects.withStroke(linewidth=2.5, foreground='black', alpha=0.9)])
plt.xscale('log')
plt.xlabel('Gamma-ray Energy (eV)')
plt.ylabel('Distance from Sun (kpc)')
plt.axvline(2e15, color='black', ls='-.', lw=2.0, alpha=0.7)
plt.axvline(1.5e14, color='black', ls='-.', lw=2.0, alpha=0.7)
plt.axhline(8.5, color='black', ls=':', lw=2.0, alpha=0.8)
cbar = fig.colorbar(im1, fraction=0.05, pad=0.04)
cbar.ax.tick_params(labelsize=14)
```

```python
energies, p_vs_e = atlas.get_survival_vs_energy(target_l, target_d, target_d)

plt.plot(energies, p_vs_e, color='red', lw=2.0, label='GPSP R12')
plt.xscale('log')
plt.ylim(0.2, 1.0)
plt.xlim(1e12, 1e16)
plt.xlabel('Gamma-ray Energy (eV)')
plt.ylabel('Survival Probability')
plt.legend(frameon=False, loc='lower left')
```

```python
distances, p_vs_d = atlas.get_survival_vs_distance(target_l, target_b, target_e)
distances2, p_vs_d2 = atlas.get_survival_vs_distance(target_l, target_b, target_e2)

plt.plot(distances, p_vs_d, color='red', lw=2.0, label='GPSP R12 (2 PeV)')
plt.plot(distances2, p_vs_d2, color='red', lw=2.0, label='GPSP R12 (150 TeV)', ls='--')
plt.ylim(0, 1.00)
plt.xlim(0, 20.0)    
plt.xlabel('Distance from Sun (kpc)')
plt.ylabel('Survival Probability')
plt.legend(frameon=False)  
        
```
